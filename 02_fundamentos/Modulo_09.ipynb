## 3. Binarización de Variables

```python
import statsmodels.formula.api as smf
import pandas as pd
```

```python
df = pd.read_csv('data/Credit.csv')
```

```python
df.dtypes
```

```python
m1 = smf.ols('Balance ~ Student', df)
fitted_m1 = m1.fit()
```

```python
fitted_m1.summary()
```

```python
import statsmodels.api as sm

Y = df.Balance.values
X = df.Student.map({'No':0, 'Yes':1}).values
```

```python
X = sm.add_constant(X)
```

```python
m2 = sm.OLS(Y,X)
fitted_m2 = m2.fit()
```

```python
fitted_m2.summary()
```
